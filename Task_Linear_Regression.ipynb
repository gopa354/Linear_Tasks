```python
#Using linear regression predicting USA house prices
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("/home/gpurama/Downloads/USA_Housing.csv")
x=df.as_matrix()
X=x[:,0]
Y=x[:,5]
print(Y.shape)

plt.scatter(X,Y)
plt.show()
#apply the equations we learned to calculate a  and b
denominator=X.dot(X)-X.mean()*X.sum()
a=(X.dot(Y)-Y.mean()*X.sum())/denominator
b=(Y.mean()*X.dot(X)-X.mean()*X.dot(Y))/denominator

#calculate predicted y
yhat=a*X+b
# let's plot the data to see what it looks like
plt.scatter(X,Y)
plt.plot(X,yhat)
plt.show()

# calculate r-squared 
d1=Y-yhat
d2=Y-Y.mean()
R2=1-d1.dot(d1)/d2.dot(d2)
print("r-squared is :",R2)
```

```python
#Using Multi linear regression predicting USA house prices
import numpy as np 
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

df=pd.read_csv("/home/gpurama/Downloads/USA_Housing.csv")
x1=df['Avg. Area House Age']
x2=df['Avg. Area Income']
Y=df['Price']
#print(Y)
X=pd.concat([x1,x2],axis=1)

X=np.array(X)
#print(X)
Y=np.array(Y)
# let's plot the data to see what it looks like
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.plot(X[:,0],Y)
plt.show()
# apply the equations we learned to calculate a and b
# numpy has a special method for solving Ax = b
# so we don't use x = inv(A)*b
# note: the * operator does element-by-element multiplication in numpy
#       np.dot() does what we expect for matrix multiplication
# = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
w=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
Yhat = np.dot(X, w)
print("Predicted Vlaues are :---",Yhat)
# determine how good the model is by computing the r-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
R2 = 1 - d1.dot(d1) / d2.dot(d2)
print("The r-squared is:", R2)
#Taking input from key board
rows = int(input("Enter number of rows in the matrix: "))
columns = int(input("Enter number of columns in the matrix: "))
x = []
print("Enter the %s x %s matrix: "% (rows, columns))
for i in range(rows):
    x.append(list(map(int, input().rstrip().split())))
x=np.array(x)
y=x.dot(w)
print(y)
```
